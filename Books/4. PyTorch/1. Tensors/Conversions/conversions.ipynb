# Interoperability: PyTorch ↔ NumPy

The integration between PyTorch and NumPy is one of the pillars of AI development. NumPy is excellent for on-CPU data processing, while PyTorch is required for GPU training and gradient computation.

---

## 1. Comparison of Resources

| Resource | NumPy ('ndarray') | PyTorch ('Tensor') |
| :- | :- | :- |
| **Hardware** | CPU only | CPU, GPU (CUDA) and MPS (Mac) |
| **Gradients** | Does not have (manual) | Native via **Autograd** |
| **Focus** | Science and General Mathematics | Deep Learning and Computer Vision |
| **Performance** | High CPU | Very high GPU (parallelism) |

## 2. Discover the type of the vector or matriz

`.type()`

## 3. Transform

* **Numpy -> Torch**: `torch.from_numpy(X)`
* **Torch -> Numpy**: `.numpy()`
    1. `.deatch()`: Disable the calculus graph
    2. `.cpu()`: Moves the GPU for CPU

```python
import torch
import numpy as np

tensor = torch.tensor([1,2,3])
numpy = np.array([1,2,3])

print(type(tensor))
print(type(numpy))
```

```python
tensor_transformed = tensor.detach().cpu().numpy()
numpy_transformed = torch.from_numpy(numpy)

print(type(tensor_transformed))
print(type(numpy_transformed))
```
