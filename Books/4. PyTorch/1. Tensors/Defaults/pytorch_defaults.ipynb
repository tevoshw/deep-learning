# Global Dtype Management in PyTorch (INT E FLOAT)

In PyTorch, the default for floating point tensors is 'torch.float32'. You can manage this configuration globally using the 'set_default_dtype' and 'get_default_dtype' functions.

---

## Control Functions

### 1. `torch.set_default_dtype(d)`
This function changes the default data type to **floating point tensors**. 
### 2. `torch.get_default_dtype()`
Returns the actual default dtype

> [!IMPORTANT]
> **What it affects:** Only decimal type conversion (float -> float).
> **What it DOES NOT affect:** Integer or boolean types remain with their fixed defaults to ensure the integrity of indexes and masks (int -> int).


```python
import torch

tensor1 = torch.tensor([1.2,2.1,3])
print(tensor1.dtype)
print(torch.get_default_dtype())
```

```python
torch.set_default_dtype(torch.float32)
torch2 = torch.tensor([1.1,2.2,3])
print(torch2.dtype)
print(torch.get_default_dtype())
```

```python
torch3 = torch.tensor([1,2,3])
print(torch3.dtype)
print(torch.get_default_dtype())
```

# Global Dtype Management (DEVICE)

In PyTorch, the default for the device 'cpu'. You can manage this configuration globally using the 'set_default_device' and 'get_default_device' functions.

---

## Control Functions

### 1. `torch.set_default_device(d)`
This function defines what the default data type will be for **new floating point tensors**. 

### 2. `torch.get_default_device()`
Returns the actual default device

### 3. `torch.cuda.is_available()`
See if cuda it's available for the device

> [!IMPORTANT]
> **Float Behavior -> Float:** If you set 'float64', any decimal number (e.g. '1.0') will be created with that precision.
> **Int Behavior -> Int:** This function **does not affect integers**. Integers (e.g. '1') will continue to be created as 'torch.int64' (Long) to ensure the stability of system indexes and counters.

```python
print(torch.get_default_device())
```

```python
torch.set_default_device('cuda')
```

```python
print(torch.get_default_device())
```

```python
print(torch.cuda.is_available())
```

# Global Gradient Management (AUTOGRAD)

In PyTorch, **Autograd** is the engine that enables the training of neural networks. It works by recording all operations performed on tensors that have tracking enabled, allowing the automatic calculation of derivatives (gradients).

When active, the computer weighs more, as it stores the computer graphs, when turned off it tends to be lighter for the computer

---

## Control Functions

### 1. `tensor = torch.tensor(..., requirements_grad=True)`
Individually defines that a tensor must "write down" its operations in a logbook. It is essential for the **weights (weights)** of a model, as it allows you to calculate how much they must change to reduce the error.

### 2. `torch.set_grad_enabled(mode)`
Enables or disables the gradient recording mechanism **globally**. 
* **True:** (Default) PyTorch tracks operations to allow '.backward()'.
* **False:** PyTorch bypasses tracing, saving memory and processing.

### 3. `torch.is_grad_enabled()`
Returns a boolean checking if gradient tracing is currently active in the global scope.

### 4. `with torch.no_grad():`
A context manager used to temporarily disable gradient in a specific block of code. It is the most common way to perform **inference** (prediction) efficiently.



---

## The Tensor's "Logbook"

When 'requires_grad=True' is active, the tensor gains special attributes to manage its mathematical memory:

* **'grad_fn'**: Keeps track of the last operation (e.g. if the tensor was born from a sum, 'grad_fn' will be 'AddBackward').
* **'grad'**: Where the gradient value (the "adjustment direction") is stored after we run the '.backward()' command.

> [!IMPORTANT]
> **Chain Tracking:** If you perform an operation between a tensor that has 'requires_grad=True' and one that does not, the result will **automatically** inherit 'requires_grad=True'.
> **VRAM Consumption:** Keeping gradient enabled consumes much more video memory, as PyTorch needs to store the "map" (graph) of all accounts to be able to go back and undo them later.



```python
print(torch.set_grad_enabled(False))
```

```python
print(torch.is_grad_enabled())
```
