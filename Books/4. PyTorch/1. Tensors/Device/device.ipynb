# Device

**Refers to the place of the data and the model is processed, and can be:**
1. CPU (Slower)
2. GPU (Fast)

### 0. `.device` See the actual device of the tensor/model
### 1. `device = torch.device('X')` or `torch.device('X')` Change the device, the X it's a string
### 2. `X = X.to(device=device)` or `X = X.to(device = torch.device('X'))` Change the things in pytorch
- Don't have inplace.

> We saw in /Defaults how to put one of them as a default for not do this evertime

```python
import torch

tensorX = torch.tensor([1,2.3])
print(tensorX.device)
```

```python
device = torch.device('cuda')

tensorX = tensorX.to(device)

print(tensorX.device)
```

```python
tensorY = torch.tensor([1,2,3]).to(torch.device('cuda'))
print(tensorY.device)
```
