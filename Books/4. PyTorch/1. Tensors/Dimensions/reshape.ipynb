# Reshape

**Returns a tensor with the same data and number of elements as input, but with the specified shape. When possible, the returned tensor will be a view of input. Otherwise, it will be a copy. Contiguous inputs and inputs with compatible strides can be reshaped without copying, but you should not depend on the copying vs. viewing behavior**

### `tensorX.reshape`(a,b): Reshape the tensor dimensions





### Reshape (-1)

**The -1 in reshape means:**

- "Automatically fills in this value to fit perfectly." You choose one dimension, and Torch separates the other equally to fit

```python
import torch

tensor1 = torch.randint(1, 10, size=(4,4))
print(f"Shape of original tensor: {tensor1.shape}")
print(f'Tensor:{tensor1}\n\n')

tensor_reshaped = tensor1.reshape(-1, 1)
print(f"Shape of reshaped tensor: {tensor_reshaped.shape}")
print(f'Tensor:{tensor_reshaped}\n\n')

tensor_reshaped2 = tensor1.reshape(1, -1)
print(f"Shape of reshaped tensor: {tensor_reshaped2.shape}")
print(f'Tensor:{tensor_reshaped2}\n\n')


```

```python
tensor_reshaped3 = tensor1.reshape(-1, 2) # Divide the columns in 2, and the rows you decided
print(f"Shape of reshaped tensor: {tensor_reshaped3.shape}")
print(f'Tensor:{tensor_reshaped3}')


""" Here it's possible split cuz 16 into 2 columns will give a integer number (16/2 = 8)"""
```

```python
tensor_reshaped4 = tensor1.reshape(-1, 3) 
print(f"Shape of reshaped tensor: {tensor_reshaped4.shape}")
print(f'Tensor:{tensor_reshaped4}')


""" Here it's impossible to split equaly 16 numbers into a 3 columns (16/3 = 5,1....) """ 
```

### Reshape Normal  

```python
import torch

tensor1 = torch.randint(1, 10, size=(4,4))
print(f"Shape of original tensor: {tensor1.shape}")
print(f'Tensor:{tensor1}\n\n')

tensor_reshaped = tensor1.reshape(1,16)
print(f"Shape of reshaped tensor: {tensor_reshaped.shape}")
print(f'Tensor:{tensor_reshaped}\n\n ')

tensor_reshaped2 = tensor1.reshape(16,1)
print(f"Shape of reshaped tensor: {tensor_reshaped2.shape}")
print(f'Tensor:{tensor_reshaped2}')
```

```python
tensor_reshaped4 = tensor1.reshape(2, 8)
print(f"Shape of reshaped tensor: {tensor_reshaped4.shape}")
print(f'Tensor:{tensor_reshaped4}')
```

```python
tensor_reshaped5 = tensor1.reshape(8,2)
print(f"Shape of reshaped tensor: {tensor_reshaped5.shape}")
print(f'Tensor:{tensor_reshaped5}')
```
