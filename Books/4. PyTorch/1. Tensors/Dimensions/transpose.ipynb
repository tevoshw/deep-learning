# Transpose

**Returns a tensor with its dimensions swapped. The data remains the same, but the order of the axes is changed. The returned tensor is a view of the original tensor (no data copy), and it may result in a non-contiguous tensor.**

### `tensorX.transpose(dim0, dim1)`: Swap two dimensions of the tensor

```python
import torch

tensor = torch.randint(1,10, size=(2,3))
print(f'Shape before the transpose: {tensor.shape}\n\n')
print(f'Shape after the transpose: {tensor.T.shape}')
```

```python
print(f'Shape before the transpose: {tensor}\n\n')
print(f'Shape after the transpose: {tensor.T}')
```
