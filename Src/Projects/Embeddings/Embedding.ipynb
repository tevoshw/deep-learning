# Here we gonna transform text into numbers :)

## 1. Get the data

```python
from datasets import load_dataset

dataset = load_dataset("imdb")

print(f'The dataset: {dataset["train"][0]['text']}')
```

## 2. Define the hyperparameters

```python
"""
PART I: Tokenizer

1. Here we need to create an index for each word
    - The movie was great -> 0, 1, 2, 3
2. All words are put in lower case, for same words in differents written be the same (this depende of the version that you get of the lib or the logic)
    - "the" and "The" will be the same = "the", "the"ArithmeticError
3. Add 2 special tokens for the start [CLS] and for the end [SEP]

"""
from transformers import AutoTokenizer
import torch

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text = dataset["train"][0]['text']
tokens = tokenizer(text)
indexs = torch.tensor(tokens['input_ids'])

"""
tokens will return


# {
#   'input_ids':      [101, 1996, 3185, 2001, 2307, 102],
#   'token_type_ids': [0, 0, 0, 0, 0, 0],
#   'attention_mask': [1, 1, 1, 1, 1, 1]
# }

"""


print(f'Tokens in their natural way (words): {dataset['train'][0]['text']}') # other way to see the tokens in words way after the tokenization -> : print(tokenizer.convert_ids_to_tokens(tokens["input_ids"]))
print(f'Tokens in their transformerd way (tokens): {indexs}')
```

```python
"""
PART II: Vocab Size


1. Here we need to know the size of vocab, (how many differentes words, signals and special characters exist in our dataset)
    - In this case how we gonna use the BERT tokenizer, the that will return it's the differents words that BERT know, and not the words our text have!
"""


vocab_size = tokenizer.vocab_size
print(vocab_size)  
```

```python
"""
PART III: Embedding dimensions


1. Here we need define how much information our words have
    - Each dim, it's a characteristic of the word
"""

embedding_dim = 64
```

```python
"""
PART IV: Embedding function


1. Here we deffine the embedding function
"""

import torch.nn as nn 

embedding = nn.Embedding(
    num_embeddings=vocab_size,
    embedding_dim=embedding_dim
)
```

```python
"""
PART V: Embedding the tokens


1. Here the indexs will be a sparse matriz with the characteristics that we define in the embedding function

"""

output = embedding(indexs)
print(f'What output will return?: {output.shape}')
# In this case output will return 363 tokens, and each token with 64 characteristics


print(f'Original text: {dataset['train'][0]['text']}')
print(f'First word of the text in their original way (words): `{dataset['train'][0]['text'][:1]}` ')
print(f'First word of the text in their TRANSFORMED way (numbers):  {output[0]}')
print(f'Len of first word of the text: {output[0].shape}')
```
